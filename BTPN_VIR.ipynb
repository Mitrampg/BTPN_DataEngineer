```python
%%sql
SELECT 
    p.CLIENTNUM,
    p.Customer_Age,
    p.Gender,
    p.Dependent_count,
    p.Income_Category,
    p.Months_on_book,
    p.Total_Relationship_Count,
    p.Months_Inactive_12_mon,
    p.Contacts_Count_12_mon,
    p.Credit_Limit,
    p.Total_Revolving_Bal,
    p.Avg_Open_To_Buy,
    p.Total_Trans_Amt,
    p.Total_Trans_Ct,
    p.Avg_Utilization_Ratio,
    c.Card_Category,
    e.Education_Level,
    m.Marital_Status,
    s.status AS Customer_Status
FROM 
    customer_data_history p
JOIN 
    category_db_2 c ON p.card_categoryid = c.id
JOIN 
    education_db_2 e ON p.Educationid = e.id
JOIN 
    marital_db_2 m ON p.Maritalid = m.id
JOIN 
    status_db_2 s ON p.idstatus = s.id
    
LIMIT 20;
```
